```python
titulos_brasileiros = ['1972', '1974', '1981', '1983', '1987', '1990', '1991']
de_2005_a_2008 = ['2005', '2006', '2007', '2008']
de_2010_a_2013 = ['2010', '2011', '2012', '2013']
de_2018_a_2021 = ['2018', '2019', '2020', '2021']
```

```python
def webscraping_pilotos_pontuacao(temporadas):
    for temporada in temporadas:
    
        url = f"https://www.statsf1.com/pt/{temporada}.aspx"
        option = Options()
        option.headless = True
        driver = webdriver.Firefox()
        driver.get(url)
        driver.find_element_by_xpath("//div[@class='SimpleBar']//table//tbody//tr//td[@class='Total']").click()
        element = driver.find_element_by_xpath("//div[@class='SimpleBar']//table")
        html_content = element.get_attribute('outerHTML')
        soup = BeautifulSoup(html_content, 'html.parser')
        table = soup.find(name='table')
        df_full = pd.read_html( str(table) )[0]
        df_full.columns = df_full.iloc[0]
        df_full.drop(df_full.index[0])
        df_full.to_csv(f'./dados_pilotos_pontuacao/pontuacao_pilotos_temporada_{temporada}.csv', index=False, encoding='utf-8')
```

```python
webscraping_pilotos_pontuacao(de_2005_a_2008)
```

# Tabela de resultado de acordo com a posição por corrida

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/2005_Formula_One_World_Championship'

response = urlopen(url)
html = response.read()

soup = BeautifulSoup(html, 'html.parser')
soup
```

```python
table = soup.find('span', id = "World_Drivers'_Championship_standings").find_next('td')
```

```python
df_full = pd.read_html( str(table) )[0]
```

```python
df_full
```

```python
def webscraping_pilotos_resultado(temporadas):
    for temporada in temporadas:
        url = f'https://en.wikipedia.org/wiki/{temporada}_Formula_One_World_Championship'

        response = urlopen(url)
        html = response.read()
        soup = BeautifulSoup(html, 'html.parser')
        table = soup.find('span', id = "World_Drivers'_Championship_standings").find_next('td')
        dados = pd.read_html( str(table) )[0]
        dados.to_csv(f'./dados_pilotos_resultados/resultado_pilotos_temporada_{temporada}.csv', index=False, encoding='utf-8')
    
```

```python
webscraping_pilotos_resultado(de_2005_a_2008)
```
